Identificar propiedades.
$\rho_{ref} = \rho_{agua} = 1000\frac{kg}{m^3}$ \\
$\rho_{aceite} = 0.92\rho_{ref} = 920\frac{kg}{m^3}$ \\
$\rho_{f} = \rho_{r}\rho_{ref} = 13600\frac{kg}{m^3}$

---
Establecer puntos de equilibrio de presiones.
$P_{1''} = P_{2''}$

---
$P_{1''} = P_{1'} + \rho_{Agua}gh$ \\
$P_{1'} = P_{1} + \rho_{Agua}gy$ \\
$\implies P_{1''} = P_{1} + \rho_{Agua}gy + \rho_{Agua}gh$

---
$P_{2''} = P_{2'} + \rho_{f}gh$ \\
$P_{2'} = P_{2} + \rho_{Agua}gy$ \\
$\implies P_{2''} = P_{2} + \rho_{Agua}gy + \rho_{f}gh$

---

$ \therefore P_{1} + \rho_{Agua}gy + \rho_{Agua}gh = P_{2} + \rho_{Agua}gy + \rho_{f}gh $

---

$\Delta P = P_{1} - P_{2} = \rho_{f}gh - \rho_{Agua}gh = (\rho_{f}- \rho_{Agua})gh ≈ 21.98\;kPa$



```python
# @title
rho_ac = 920
rho_ag = 1000
rho_f = 13600

g = 9.81
h = 17.78*10**(-2)

DP = (rho_f - rho_ag)*g*h
print("Delta P:",round(DP/1000,2),"kPa")
```

###(2.0p) Despejar el valor de la viscosidad:
$R = \frac{D}{2}$ \\
$\bar{v} = \frac{\frac{\Delta P}{4 L \mu_{aceite}} \frac{D^2}{4}}{2} = \frac{\Delta P D^2}{32\;L \mu_{aceite}}$ \\
$\therefore \mu_{aceite} = \frac{\Delta P D^2}{32\;L\;\bar{v} }$

---
###(1.0p) Obtener la viscosidad:

$\therefore \mu_{aceite} = \frac{\Delta P D^2}{32\;L\;\bar{v} } = \frac{21.98\times 10^{3}\cdotp2.54\times10^{-3}}{32\cdotp30.48\times10^{-2}\cdotp 1.47} ≈ 0.01 Pa\cdotp s$

---

```python
# @title
D = 2.54*10**(-3)
L = 30.48*10**(-2)
vp = 1.47

mu = (DP*D**2)/(32*L*vp)
print("viscosidad, mu:", round(mu,3), "Pa s")
```
