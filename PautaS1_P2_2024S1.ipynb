###(5.0p) Establecer conservacion de masa:
$\frac{d}{dt}\int_{\forall} \rho d\forall + \int \rho(\vec{V}\cdot\hat{n})dA = 0$ \\
$\int \rho(\vec{V}\cdot\hat{n})dA = 0$ \\

$Q_{1} + Q_{2} = 0$ \\
$-V_{1}A_{1} + V_{2}A_{2} = 0$ \\

Con:
$A_{1} = \frac{\pi D_{1}^2}{4} = 0.00785\;m^{2}$ \; $A_{2} = \frac{\pi D_{2}^2}{4} = 0.00283\;m^{2}$

---

```python
import numpy as np

D1 = 10*10**(-2)
A1 = np.pi*D1**2/4.0
print("Area 1:",round(A1,5),"m2")

D2 = 6*10**(-2)
A2 = np.pi*D2**2/4.0
print("Area 2:",round(A2,5),"m2")
```


###(5.0p) Establecer conservacion de momento:

$\frac{d}{dt}\int_{\forall} \rho\vec{V} d\forall + \int \rho\vec{V}(\vec{V}\cdot\hat{n})dA = \sum \vec{F}_{ext}$ \\
$\int \rho(\vec{V}\cdot\hat{n})dA = \sum \vec{F}_{ext}$ \\

___

Fuerzas Externas:
$\sum \vec{F}_{ext} = \vec{F}_{p_{1}} + \vec{F}_{p_{2}} + \vec{F}_{R}$

___
Identificación de Vectores velocidad: \\

$\vec{V}_{1} = \{V_{1},0,0\}$ \\
$\hat{n}_{1} = \{-1,0,0\}$ \\
$\vec{V}_{2} = \{V_{2},0,0\}$ \\
$\hat{n}_{2} = \{1,0,0\}$ \\
___

Identificación de Flujos: \\
$\vec{V_{1}}\cdot\hat{n_{1}} = -V_{1}$ \\
$\vec{V_{2}}\cdot\hat{n_{2}} = +V_{2}$

___

Identificación de Vectores Fuerzas: \\
$\vec{F}_{p_1} = \{P_{1}A_{1},0,0\}$ \\
$\vec{F}_{R} = \{-F_{x},0,0\}$ \\

___

$ \dot{m}_{1}\vec{V}_{1} + \dot{m}_{2}\vec{V}_{2} = \vec{F}_{p_{1}} + \vec{F}_{R}$ \\
###(2.0p) Obtener ec. de equilibrio en el eje x:
$ -\rho V^2_{1}A_{1} + \rho V^2_{2}A_{2} = P_{1}A_{1} - F_{x}$ \\

###(5.0p) Resolver el sistema de ecuaciones:

$-V_{1}A_{1} + V_{2}A_{2} = 0$ \\
$ -\rho V^2_{1}A_{1} + \rho V^2_{2}A_{2} = P_{1}A_{1} - F_{x}$ \\

Despejando $V_{2}$: \\

$V_{2} = \sqrt{\frac{P_{1}A_{1} - F_{x}}{\rho \left( A_{2} - \frac{A^{2}_{2}}{A_{1}} \right)}} = 30\frac{m}{s}$


```python
P1 = 321300
Fx = 900
rho = 1000

V2 = np.sqrt((P1*A1 - Fx)/(rho*(A2- A2**2/A1)))
print("Velocidad en 2:", round(V2,1), "m/s")
```


###(1.0p) Despejar el caudal:
$Q = V_1A_{1} = V_{2}A_{2} ≈ 0.085\frac{m^3}{s} $

```python
Q = V2*A2
print("Caudal: ", round(Q,3),"m3/s")
```
